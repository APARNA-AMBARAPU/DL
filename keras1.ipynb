```
import numpy as np
import matplotlib.pyplot as plt
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

from sklearn import datasets
```

```
n_pts = 500
np.random.seed(0)
Xa = np.array([np.random.normal(13, 2, n_pts),
               np.random.normal(12, 2, n_pts)]).T
Xb = np.array([np.random.normal(8, 2, n_pts),
               np.random.normal(6, 2, n_pts)]).T
 
X = np.vstack((Xa, Xb))
y = np.matrix(np.append(np.zeros(n_pts), np.ones(n_pts))).T
 
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
```

```
npoints = 500
Xa  = np.array([np.random.normal(13,2,npoints), np.random.normal(12,2,npoints)]).T
Xb = np.array([np.random.normal(8,2,npoints),np.random.normal(6,2,npoints)]).T
```

```
y = np.asarray([np.ones(npoints),np.zeros(npoints)]).reshape(2*npoints,1)
```

```
X = np.vstack((Xa,Xb))
```

```
plt.scatter(X[:npoints,0],X[:npoints,1])
plt.scatter(X[npoints:,0],X[npoints:,1])
```

```
X.shape
```

```
y.shape
```

```
model = Sequential()
```

```
model.add(Dense(units=1,input_shape = (2,),activation = 'sigmoid'))
```

```
model.compile(Adam(0.01),'binary_crossentropy',metrics = ['accuracy'])
```

```
h = model.fit(X,y,epochs = 100,batch_size=20,verbose = 1,shuffle = True)
```

```
 accuracy = h.history['accuracy']
```

```
loss = h.history['loss']
```

```
plt.plot(range(1,len(accuracy)+1),accuracy)
#plt.plot(range(1,101),loss)
```

```
def plot_decision_boundary(X,y,model):
  xspan = np.linspace(X[:,0].min() -1,X[:,0].max() + 1)
  yspan = np.linspace(X[:,1].min() - 1,X[:,1].max() + 1)
  xx,yy = np.meshgrid(xspan,yspan)
  xx_,yy_ = xx.ravel(),yy.ravel()
  grid = np.c_[xx_,yy_]
  predictions = model.predict(grid)
  z = predictions.reshape(xx.shape)
  plt.contourf(xx,yy,z)
```

```
plot_decision_boundary(X,y,model)
plt.scatter(X[:npoints,0],X[:npoints,1])
plt.scatter(X[npoints:,0],X[npoints:,1])
```

```
point = np.array([[2.5,5.0]])
```
